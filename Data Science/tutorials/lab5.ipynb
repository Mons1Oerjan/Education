```python
import pandas as pd
import numpy as np
```

```python
# source of data :https://www.kaggle.com/vikrishnan/boston-house-prices
df = pd.read_csv('housing_clean.csv')
```

```python
df.head()
```

### Remove NaN values

```python
df.dropna(inplace=True)
```

### Divide dataset into training data and target / test data

```python
y = df.iloc[:,-1:].values  # MEDV
x = df.iloc[:,:-1].values  # All other columns
```

### Standardize features by removing the mean and scaling to unit variance

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(x)
```

```python
# divide data to test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=.4,
    random_state=42
)
```

### Define a classifier (Random Forest)

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
clf = RandomForestClassifier(
    max_depth=5,
    n_estimators=10,
    max_features=1
)
```

### Define a Least Squares Linear Regression Model

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()
```

### Fit the linear regression model

```python
lr.fit(X_train, y_train.ravel())
```

### Predict and calculate the mean error of the test set and train set

```python
from sklearn.metrics import mean_absolute_error

# Test set:
y_pred = lr.predict(X_test)
# list(zip(y_pred,y_test.ravel()))
mae = mean_absolute_error(y_test.ravel(), y_pred)
print('mean absolute error for test set :', mae)

# train set:
y_pred = lr.predict(X_train)
mae = mean_absolute_error(y_train.ravel(), y_pred)
print('mean absolute error for train set :', mae)
```

### Define a Least Angle Regression Model

```python
from sklearn import linear_model
reg = linear_model.Lars(n_nonzero_coefs=4)
reg.fit(X_train, y_train.ravel())
```

### Predict and calculate mean error of the test set and train set

```python
# Test set:
y_pred = reg.predict(X_test)
mae = mean_absolute_error(y_test.ravel(), y_pred)
print('mean absolute error for test set :', mae)

# Train set:
y_pred = reg.predict(X_train)
mae = mean_absolute_error(y_train.ravel(), y_pred)
print('mean absolute error for train set :', mae)
```

### Define a Multilayer Perceptron Regressor Model

```python
from sklearn.neural_network import MLPRegressor
mlp = MLPRegressor(
    hidden_layer_sizes=(60,),
    max_iter=100,
    alpha=1e-4,
    verbose=10,
    tol=1e-8,
    random_state=4,
    learning_rate_init=.001
)
```

### Fit the MLP Regressor Model

```python
mlp.fit(X_train, y_train.ravel())
```

### Predict and calculate the mean error of the test set and train set

```python
# Test set:
y_pred = mlp.predict(X_test)
mae = mean_absolute_error(y_test.ravel(), y_pred)
print('mean absolute error for test set :', mae)

# Train set:
y_pred = mlp.predict(X_train)
mae = mean_absolute_error(y_train.ravel(), y_pred)
print('mean absolute error for train set :', mae)
```

### Perform a GridSearch on a Kernel Ridge Regression
Kernel ridge regression (KRR) combines ridge regression (linear least squares with l2-norm regularization) with the kernel trick. It thus learns a linear function in the space induced by the respective kernel and the data. For non-linear kernels, this corresponds to a non-linear function in the original space.

```python
from sklearn.kernel_ridge import KernelRidge
from sklearn.model_selection import GridSearchCV
kr = GridSearchCV(
    KernelRidge(kernel='rbf', gamma=0.1),
    cv=5,
    param_grid={
        "alpha": [1e0, 0.1, 1e-2, 1e-3],
        "gamma": np.logspace(-2, 2, 5)
    }
)
```

### Fit the KRR Model

```python
import time
t0 = time.time()
kr.fit(X_train, y_train)
kr_fit = time.time() - t0
print("Kernel ridge model fitted in %.3f s"% kr_fit)
```

### Predict and calculate the mean error of the train and test sets

```python
# Test set:
y_pred = kr.predict(X_test)
mae = mean_absolute_error(y_test.ravel(), y_pred)
print('mean absolute error for test set :', mae)

# Train set:
y_pred = kr.predict(X_train)
mae = mean_absolute_error(y_train.ravel(), y_pred)
print('mean absolute error for train set :', mae)
```
