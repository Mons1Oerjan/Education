```python
df = pd.read_csv('1.csv', delimiter=',', parse_dates=['collected_time'])
df.sort_index()
df.sort_values(by=['collected_time'], inplace=True)
df.set_index('collected_time', inplace=True)
```

```python
df.head()
```

```python
speed = df.loc[:, 'speed']
print('mean: ', speed.mean())
print('std dev: ', speed.std())
```

```python
td = df.loc[:, 'td']
print('mean: ', td.mean())
print('std dev: ', td.std())
```

```python
i = 0
seven_points = df.iloc[i:i+7, :]
latitude = seven_points.lat.values
longitude = seven_points.lon.values
mid = df.iloc[4:5, :]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(latitude, longitude)

plt.scatter(mid.lat, mid.lon, c='g')

mid_lat = df.iloc[3:4, :].lat + df.iloc[5:6, :].lat / 2
mid_lon = df.iloc[3:4, :].lon + df.iloc[5:6, :].lon / 2
plt.scatter(mid_lat, mid_lon, c='red')

lat_min = latitude.min()
lat_max = latitude.max()
lon_min = longitude.min()
lon_max = longitude.max()
plt.xlim(lat_min-0.00001, lat_max+0.00001)
plt.ylim(lon_min-0.00001, lon_max+0.00001)

plt.show()
```
